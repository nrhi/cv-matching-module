```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r"\s+", " ", text)
    text = re.sub(r"[^a-zA-ZÀ-ÿ0-9\s+#./-]", "", text)
    return text.strip()
```

```python
with open("../outputs/cv_text.txt", "r", encoding="utf-8") as f:
    raw_text = f.read()

cleaned_text = clean_text(raw_text)

print(cleaned_text[:2000])
```

```python
with open("../outputs/cv_cleaned.txt", "w", encoding="utf-8") as f:
    f.write(cleaned_text)

print("Cleaned text saved successfully!")
```

```python

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

documents = [cleaned_text] + job_descriptions

vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(documents)

cv_vector = tfidf_matrix[0:1]
job_vectors = tfidf_matrix[1:]

scores = cosine_similarity(cv_vector, job_vectors)

print(scores)
for i, score in enumerate(scores[0], start=1):
    print(f"Job {i}: {score:.4f}")
```

```python
job_descriptions = [
    "Nous recherchons une étudiante avec un bon sens relationnel, le travail en équipe, l'accueil et le service client pour un poste d'hôtesse d'accueil dans un hôtel.",
    
    "Nous recrutons un développeur Python avec Django, SQL, API et expérience en machine learning.",
    
    "Stage en traduction et langues étrangères. Le candidat doit parler français, anglais et italien et avoir de bonnes compétences en communication."
]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

documents = [cleaned_text] + job_descriptions

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)

cv_vector = tfidf_matrix[0:1]
job_vectors = tfidf_matrix[1:]

scores = cosine_similarity(cv_vector, job_vectors)

for i, score in enumerate(scores[0], start=1):
    print(f"Job {i}: {score:.4f}")
```

```python
feature_names = vectorizer.get_feature_names_out()
print(feature_names[:100])
```

```python
with open("../outputs/cosine_scores.txt", "w", encoding="utf-8") as f:
    for i, score in enumerate(scores[0], start=1):
        f.write(f"Job {i}: {score:.4f}\n")

print("Scores saved successfully!")
```

```python
important_words = [
    "python", "django", "sql", "machine", "learning",
    "anglais", "français", "italien", "traduction",
    "communication", "service", "client", "langues"
]
```

```python
def filter_important(text, keywords):
    words = text.split()
    filtered = [word for word in words if word in keywords]
    return " ".join(filtered)

cv_filtered = filter_important(cleaned_text, important_words)

print(cv_filtered)
```

```python
filtered_jobs = [filter_important(job.lower(), important_words) for job in job_descriptions]

print(filtered_jobs)
```

```python
documents = [cv_filtered] + filtered_jobs

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)

cv_vector = tfidf_matrix[0:1]
job_vectors = tfidf_matrix[1:]

scores = cosine_similarity(cv_vector, job_vectors)

for i, score in enumerate(scores[0], start=1):
    print(f"Job {i}: {score:.4f}")
```

```python
job_scores = list(enumerate(scores[0], start=1))

# sort from highest to lowest
job_scores_sorted = sorted(job_scores, key=lambda x: x[1], reverse=True)

for job_id, score in job_scores_sorted:
    print(f"Job {job_id}: {score:.4f}")
```
